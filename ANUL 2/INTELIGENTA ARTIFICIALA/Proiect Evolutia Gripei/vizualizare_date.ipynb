### Citire date din fisierul [fluprint_export](./fluprint_export.csv)

```python
def read_from_fluprint(colsInCommon) -> pd.DataFrame:
    df = pd.read_csv('fluprint_export.csv')
    df = df.rename(columns={'donor_id':'DonorID'})
    df_unique = df[['DonorID', 'vaccine_response']].drop_duplicates()
    df_unique = df_unique.rename(columns={'vaccine_response':'outcome'})
    pivot_df = df.pivot_table(index='DonorID', columns='name', values='data', aggfunc='first')
    
    pivot_df.reset_index(inplace=True)
    pivot_df.fillna(0, inplace=True)
    pivot_df = pivot_df.astype({'DonorID': int})
    pivot_df = pd.merge(pivot_df, df_unique, on='DonorID', how='left')
    outcome = []
    for o in pivot_df['outcome']:
        if o == 0.0:
            outcome.append('low')
        else:
            outcome.append('high')
    pivot_df['outcome'] = outcome

    cols_to_drop = []
    for column in pivot_df.columns:
        if column.replace(' ','').lower() not in [c.replace(' ','').lower() for c in colsInCommon]:
            cols_to_drop.append(column)
    pivot_df = pivot_df.drop(columns=cols_to_drop)
    return pivot_df
```

```python
def normalizare_date(dataFrame:pd.DataFrame) -> pd.DataFrame:
    from numpy import log
    normalized_dataFrame = dataFrame.copy()

    for column_name,column_type in zip(normalized_dataFrame.columns,normalized_dataFrame.dtypes):
        if column_type in ['int64','float64']:
            maxi = normalized_dataFrame[column_name].max()
            mini = normalized_dataFrame[column_name].min()
            normalized_dataFrame[column_name] = (normalized_dataFrame[column_name] - mini )/ (maxi - mini)
    return normalized_dataFrame
```

```python
def plot_datas(df: pd.DataFrame, columns):
    nr_cols = 4
    nr_rows = len(columns)//nr_cols
    fig,axes = plt.subplots(nrows=nr_rows+1,ncols=nr_cols,figsize=(20,65))
    current_row = 0
    current_col = 0

    for column in columns:
        axes[current_row][current_col].hist(df[df['outcome'] == 'high'][column],fc=(1,0,0,0.99))
        axes[current_row][current_col].hist(df[df['outcome'] == 'low'][column],fc=(0,1,0,0.50))
        axes[current_row][current_col].set_title(column)
        current_col += 1
        if current_col == nr_cols:
            current_col = 0
            current_row += 1 
    plt.show()
```

```python
def clasificationPerformance(ground_truth, computed_values, positive_label = "high"):
    """
    Returneaza TN (True Negative), FP(False Positive), FN(False Negative), TP(True Positive)
    """
    TN = 0
    FP = 0
    FN = 0
    TP = 0
    
    for i in range(0, len(ground_truth)):
        if ground_truth[i] == positive_label:
            if computed_values[i] == positive_label:
                TP += 1
            else:
                FP += 1
        else:
            if computed_values[i] != positive_label:
                TN += 1
            else:
                FN += 1
    return TN, FP, FN, TP
```

```python
def getAccuracy(TN, FP, FN, TP):
    """ 
    accuracy represents the overall performance of classification model:
    (TP+TN)/(TN+FP+FN+TP)
    """
    if (TN+FP+FN+TP) == 0:
        return 0
    return (TP+TN)/(TN+FP+FN+TP)
```

```python
def getPrecision(FP, TP):
    """
    precision indicates how accurate the positive predictions are 
    TP/(TP+FP)
    """
    if (TP+FP) == 0:
        return 0
    return TP/(TP+FP)
```

```python
def getRecall(TP, FN):
    """ 
    recall indicates the coverage of actual positive sample
    TP/(TP+FN)
    """
    if (TP+FN) == 0:
        return 0
    return TP/(TP+FN)
```

## Impartirea datelor in date de test si date de antrenament
- 70% date de training
- 30% date de validare

```python
def getTrainingAndValidationDatas(cols, cols_in_common,file=File.FLUPRINT):
    import numpy as np
    df = read_datas(file,cols_in_common)
    n = df.shape[0]
    np.random.seed(69)
    indexes = [i for i in range(n)]
    trainingIndexes = np.random.choice(indexes, int(0.7 * n), replace = False)
    validationIndexes = [i for i in range(n) if not i in trainingIndexes]

    trainingInputs = [df[cols].iloc[i] for i in trainingIndexes]
    trainingOutputs = [df['outcome'].iloc[i] for i in trainingIndexes]
    trainingInputs = [[1] + inputs for inputs in trainingInputs]

    validationInputs = [df[cols].iloc[i] for i in validationIndexes]
    validationOutputs = [df['outcome'].iloc[i] for i in validationIndexes]
    validationInputs = [[1] + validation for validation in validationInputs]

    return trainingInputs, trainingOutputs, validationInputs, validationOutputs
```

## Antrenarea unui clasificator

### Folosind tool (neural_network din sklearn)

```python
from sklearn import neural_network

def getClassificatorUsingTool(hidden_layer, activation_function, trainInputs, trainOutputs):
    classifier = neural_network.MLPClassifier(hidden_layer_sizes=(hidden_layer,), activation=activation_function, max_iter=1000, solver='sgd', verbose=10, random_state=1, learning_rate_init=.1)
    classifier.fit(trainInputs, trainOutputs)
    return classifier
```

## Testarea clasificatorului

Functie de testare a calitatii clasificatorului

Este esentiala o precizie buna pentru "high".

```python
def test_classificator_qualitty(clasificator, validation_input, validation_output, positive_label):
    computed_output = clasificator.predict(validation_input)
    TN, FP, FN, TP = clasificationPerformance(validation_output,computed_output,positive_label)
    accuracy = getAccuracy(TN,FP,FN,TP)
    precision = getPrecision(FP,TP)
    recall = getRecall(TP,FN)
    print("\nAccuracy: {}\nPrecision: {}\nRecall: {}".format(accuracy,precision,recall))
```
